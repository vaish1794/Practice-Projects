# Encoding the categorical features

```python

#col1 is the array of all categorical variables in the dataset
col1
```

```python
#col1d is the array of categorical features/inputs, i.e without the target("Attrition")
col1d
```

```python
df_encode=pd.get_dummies(data=df[col1d],drop_first=True)
df_new=df.drop(columns=col1d,axis=1)
df_new=df_new.join(df_encode)
df_new
```

# Balancing the output variable

```python
df_new["Attrition"].value_counts()
```

```python
from sklearn.utils import resample
dfminor_output=df_new[(df_new["Attrition"]=="Yes")]
```

```python
dfmajor_output=df_new[(df_new["Attrition"]=="No")]
```

```python
dfmin_up=resample(dfminor_output,replace=True,n_samples=1233,random_state=42)
```

```python
df_up=pd.concat([dfmin_up,dfmajor_output])
```

```python
df_up
```

```python
#Splitting Features and Target from the dataset
X=df_up.drop(['Attrition'],axis=1)
Y=df_up[["Attrition"]]
```

```python
Y_e=Y.replace({'Yes': 1, 'No': 0}).astype(int)
```

```python
#Checking the Variance Inflation Factor
from statsmodels.stats.outliers_influence import variance_inflation_factor
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"]=X.columns

# calculating VIF for each feature

vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]

print(vif_data)
```

```python
#Scaling Data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
Xnew=sc.fit_transform(X)
```

```python
Xnew.mean()
```

```python
Xnew.std()
```

```python
#Applying power transform
from sklearn.preprocessing import power_transform
Xt=power_transform(Xnew)
```

```python
#Splitting data into Training and Testing Data
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score,roc_auc_score
```

```python
lr=LogisticRegression()
rf=RandomForestClassifier()
kn=KNeighborsClassifier(n_neighbors=3)
sv=SVC()
dt=DecisionTreeClassifier(criterion='entropy')
gb=GaussianNB()
lst=[lr,rf,kn,sv,dt,gb]
```

```python
model=[]
acc_score=[]
F1s=[]
rocauc=[]
random=[]


for i1 in lst:
    for i2 in range(0,10):
        x_train,x_test,y_train,y_test=train_test_split(Xt,Y_e,random_state=i2,test_size=0.33)
        i1.fit(x_train,y_train)
        pred_test=i1.predict(x_test)
        model.append(i1)
        acc_score.append(round(accuracy_score(y_test,pred_test)*100,2))
        F1s.append(round(f1_score(y_test,pred_test)*100,2))
        rocauc.append(round(roc_auc_score(y_test,pred_test)*100,2))
        random.append(i2)
#Creating Dataframe of comparison between different models and their respective scores

modeldata=pd.DataFrame({"Model":model,"Random State":random,"Accuracy Score":acc_score,"F1 Score":F1s,"Roc_Auc Score":rocauc})
modeldata     
```

Ideally the accuracy score should be between 70-90%, f1 score-1,Roc_Auc score - 1
From the above section, we have trained different models for a range of random state values and after predcting the output,
we have plotted 3 types of scores for each model.
From the above output, it is clear that Random Forest Classifier with random state-5, test_size=0.33 has better scores compared to the other models.

Accuracy Score-97.91
F1 Score-97.99
Auc_Roc score-97.88
